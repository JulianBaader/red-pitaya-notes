```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('data.npy')
for i in range(10):
    plt.plot(data[i][0], label="IN1")
    plt.plot(data[i][1], label="IN2")
    plt.legend()
    plt.show()
```

```python
maxs_1 = [np.max(data[i][0]) for i in range(data.shape[0])]
plt.hist(maxs_1,bins=np.linspace(0,4095,4096), histtype='step')
plt.show()
```
